<b>Importing the Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

<b>Read the CSV File

```python
data = pd.read_csv('stock_prices.csv')
data.head()
```

<b>Convert Date to DateTime

<b>Checking the Data info and Date Column Typ

```python
data.info()
```

```python
data['date'] = pd.to_datetime(data['date'])
```

<b>Perform Time Series Analysis and Plot

```python
rolling_average = data['closing_price'].rolling(window=30).mean()

plt.figure(figsize=(13, 6))
plt.plot(data['date'], data['closing_price'], label='Closing Price')
plt.plot(data['date'], rolling_average, label='30-Day Moving Average', color='red')
plt.legend()
plt.xlabel('Date', fontsize=12, fontweight='bold')
plt.ylabel('Price', fontsize=12, fontweight='bold')
plt.title('Stock Prices and 30-Day Moving Average\n',fontsize=16, fontweight='bold')
plt.xticks(rotation=45,fontsize=10, fontweight='bold')
plt.yticks(fontsize=10, fontweight='bold')
plt.tight_layout()
plt.show()
```
